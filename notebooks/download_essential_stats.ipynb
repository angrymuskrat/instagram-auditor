```python
with_posts = list(profiles.find({}, {
    "id": 1,
    "username": 1,
    "followedby": 1,
    "follow": 1,
    "postscount": 1,
    "posts.commentscount": 1,
    "posts.likescount": 1,
}))
```

```python
df_with_posts = pd.DataFrame(with_posts)
```

```python
df_with_posts
```

```python
print(*df_with_posts["posts"][0:2], sep='\n')
```

```python
row_prefix = lambda i: f"post_{i}" 

def expand_posts(row):    
    flattened = dict()
    for i, post in enumerate(row):
        flattened[f"{row_prefix(i)}_comments"] = post["commentscount"]
        flattened[f"{row_prefix(i)}_likes"] = post["likescount"]
    
    return pd.Series(flattened)
```

```python
posts_stats = df_with_posts["posts"].apply(expand_posts)
```

```python
posts_stats
```

```python
posts_stats.fillna(0, inplace=True)
```

```python
posts_stats
```

```python
ids = df_with_posts["id"]
posts_stats["id"] = ids
```

```python
stats = df_with_posts.merge(posts_stats)
```

```python
stats.drop(columns=["posts"], inplace=True)
```

```python
stats.to_csv("profiles.csv")
```
